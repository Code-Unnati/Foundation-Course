<span style="color:black;font-size: 25px; font-family: Arial; font-weight: bold;"> Exercise 5 - Sorting Pandas DataFrame </span>

```python
# Lets sort the rows of a DataFrame based on their index or row labels
import pandas as pd

data = pd.DataFrame({'Name': ['Alice', 'Bob', 'Charlie', 'David'],
                   'Age': [25, 50, 35, 48],
                   'Salary': [10000, 96000, 54000, 52000]})
print(data)

#Sort the rows on the basis of their index
sorted_data = data.sort_index(axis=0, ascending=False)
print("\n",sorted_data)
```

```python
# Lets sort the DataFrame based on column names,
sorted_data = data.sort_index(axis=1, ascending=True)
print(sorted_data)
```

```python
# Lets sort the rows of a dataframe based on Single column
sorted_data = data.sort_values(by='Age', ascending=True)
print(sorted_data)
```

```python
# Sorting by multiple columns
data = pd.DataFrame({'Name': ['Alice', 'Bob', 'Alice', 'David'],
                   'Age': [25, 50, 35, 48],
                   'Salary': [10000, 96000, 54000, 52000]})
print(data)

sorted_data = data.sort_values(by=['Name', 'Salary'], ascending=[True, False])
print("\n After Sorting","\n",sorted_data)
```

<span style="color:black;font-size: 25px; font-family: Arial; font-weight: bold;"> Exercise 7 - Handling Missing Values</span>

```python
# Read csv file using pandas
import pandas as pd
df = pd.read_csv("nba.csv")
df
```

```python
# Print missing values column wise
df.isnull().sum()
```

```python
# Print total missing values in dataframe
df.isnull().sum().sum()
```

#### Handling NaN values is an essential part of data analysis. Depending on the nature of your data and the analysis you want to perform, you can choose different approaches to handle missing data.
- Dropping rows which have Null values.
- Dropping column which have Null values.
- filling null values


```python
# Lets  Drop rows that have at least one missing value
clean_df=df.dropna(axis=0, how="any")
clean_df
```

```python
## Drop rows that have less than 8 non-missing values
clean_df=df.dropna(axis=0, thresh=8)
clean_df
```

```python
# Drop columns that have at least one missing value
df_clean=df.dropna(axis=1)
df_clean
```

```python
# filling missing values using fillna() method
# Lets fill missing values of Salary column
df["Salary"].isnull().sum()
```

```python
# Calculate average salary  using mean() method
avg_salary=df["Salary"].mean()
avg_salary
```

```python

df["Salary"].fillna(avg_salary,inplace=True)
```

```python
df["Salary"].isnull().sum()
```

<span style="color:black;font-size: 25px; font-family: Arial; font-weight: bold;"> Exercise 8 - Handling Duplicates Values</span>

In large datasets, we often encounter duplicate entries in tables.In most cases, duplicate data is of little use and may cause your analysis to go in the wrong direction.Pandas provides several methods to find and remove duplicate entries in DataFrames.

```python
import pandas as pd

# create dataframe
data = {
    'Name': ['Aman', 'Ram', 'Aman', 'Ram', 'Aman'],
    'Age': [27, 24, 27, 24, 19],
    'City': ['Haryana', 'Surat', 'Haryana', 'New Delhi', 'Rajkot']
}
df = pd.DataFrame(data)
print(df)
# check for duplicate entries
print("\n",df.duplicated())
```

```python
# check for duplicate entries in columns Name and Age
print(df.duplicated(subset=['Name', 'Age']))
```

```python
# remove duplicates
df.drop_duplicates(inplace=True)
df
```

```python
# Lets sort the result
result["Age"].mean().sort_values(ascending=True)
```

```python
# Lets perform multiple aggregation operation using aggregation interface
result.agg(
    avg_age = ("Age","mean"),
    coult_college=("College","count")
  )
```

<span style="color:black;font-size: 25px; font-family: Arial; font-weight: bold;"> Exercise 10 - Statistical Function in Pandas</span>

```python
# Lets demonstrate sum(), max() and median() 
import pandas as pd
# Dataset
data = {
'Maths': [90, 85, 98, 80, 55, 78],
'Science': [92, 87, 59, 64, 87, 96], 'English': [95, 94, 84, 75, 67, 65]
}

# DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print("DataFrame = \n",df)

# Display the Sum of Marks in each column 
print("\nSum = \n",df.sum())
print("\nMaximum Marks = \n", df.max())

# Display the Median of Marks in each column 
print("\nMedian = \n",df.median())

# Display the Mean of Marks in each column 
print("\nMedian = \n",df.mean())
```

```python
# Use of count() method
import pandas as pd
# Dataset
data = {
'Maths': [90, 85, 98, None, 55, 78],
'Science': [92, 87, 59, None, None, 96],
'English': [95, None, 84, 75, 67, None]
}
# DataFrame
df = pd.DataFrame(data)
# Display the DataFrame
print("DataFrame = \n", df)
# Display the Count of non-empty values in each column
print("\nCount of non-empty values = \n", df.count())
```
