###  A bar chart is a graphical representation of data where bars of uniform width are used to show the comparison of different categories or groups.

```python
# Bar Plot

# importing matplotlib module
from matplotlib import pyplot as plt

# x-axis values
x = [5, 2, 9, 4, 7]

# Y-axis values
y = [10, 5, 8, 4, 2]

# Function to plot
plt.bar(x, y)

# function to show the plot
plt.show()
```

### Horizontal bar chart


#### A bar chart can be display horizontally instead of vertically, just need to call the barh() function. But don't forget to switch the axes labels!

```python
rice = [57, 50, 47, 30]
wheat = [43, 50, 53, 70]
plt.barh(rice,wheat)
plt.grid(color='green', linestyle=':', linewidth=1.0, axis='y', alpha=0.5)

plt.xlabel('Rice', fontsize=14)
plt.ylabel('Wheat', fontsize=14)
plt.title('The results of Rice/Wheat export', fontsize=20)
plt.show()
```

### Stacked bar plot

* A stacked bar chart is a type of graph that displays multiple data points on top of each other. In a stacked bar chart, each bar represents a single category that contains smaller categories. Use it to demonstrate how parts relate to the total amount.

* In stacked bar plots need to provide the additional bottom parameter that indicates where the bar should start.# 

```python
years = ['2016', '2017', '2018', '2019']
rice = [57, 50, 47, 30]
wheat = [43, 50, 53, 70]

plt.figure(figsize=(10, 6))

plt.bar(years, rice, label='Rice')
plt.bar(years, wheat, bottom=rice, label='Wheat')

plt.xlabel('Years', fontsize=14)
plt.ylabel('Preference (%)', fontsize=14)
plt.title('The results of Rice/Wheat export', fontsize=20)
plt.legend()
plt.show()
```

* **Visualization of Relationships: Shows how two sets of data are related by plotting individual data points.**

```python
#Scatter Plot
# importing matplotlib module
from matplotlib import pyplot as plt

# x-axis values
x = [5, 2, 9, 4, 7, 3, 8, 10, 1]

# Y-axis values
y = [10, 5, 8, 4, 2, 1, 6, 3, 9]

# Function to plot scatter
plt.scatter(x, y)

# function to show the plot
plt.show()



```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x-axis values using linspace
x = np.linspace(0, 10, 30)

# Generate y-axis values using sine function
y = np.sin(x)

# Plot the data points with customized markers and lines
plt.plot(x, y, '-p',            # Solid line with circle markers
         color='gray',          # Color of the line and markers
         markersize=15,         # Size of the markers
         linewidth=4,           # Thickness of the line
         markerfacecolor='white',  # Color of the marker face
         markeredgecolor='gray',   # Color of the marker edge
         markeredgewidth=2)        # Thickness of the marker edge

# Set y-axis limits
plt.ylim(-1.2, 1.2)

# Display the plot
plt.show()
```

<span style="color:black;font-size: 25px; font-family: Arial; font-weight: bold;"> Pie Chart</span>

* A pie chart is a round graph divided into slices to show how parts of a whole compare to each other. 
* Each slice represents a category or group, and its size shows the proportion or percentage that category contributes to the total.
* They're great for visualizing simple comparisons, like market shares of different companies or the distribution of expenses in a budget.

```python
# Importing necessary libraries
from matplotlib import pyplot as plt
import numpy as np

# Step 1: Define the dataset
cars = ['AUDI', 'BMW', 'FORD', 'TESLA', 'JAGUAR', 'MERCEDES']
data = [23, 17, 35, 29, 12, 41]

# Step 2: Create a figure and set its size
fig = plt.figure(figsize=(8, 6))


# Step 3: Create a pie chart
plt.pie(data,                   
        labels=cars,             
        autopct='%1.1f%%',       # Format for percentage display
        startangle=140,          # Start angle for the first slice (140 degrees)
        colors=plt.cm.Pastel1(np.arange(len(data))),  # Colors for each slice based on Pastel1 colormap
        wedgeprops={'edgecolor': 'black', 'linewidth': 2})  # Properties for the wedge (slice) including edge color and width

# Parameters Explanation:
# - autopct='%1.1f%%': Format string for percentage values with one decimal place.
# - startangle=140: Starts the first slice at 140 degrees (anti-clockwise from the x-axis).
# - wedgeprops={'edgecolor': 'black', 'linewidth': 2}: Customizes wedge (slice) properties such as edge color and width.

# Step 4: Display the plot
plt.axis('equal')  # Ensures the pie chart is drawn as a circle
plt.title('Car Distribution Pie Chart')  # Sets the title of the plot

# Step 5: Show the plot
plt.show()
# Displays the pie chart with the specified settings.
```

```python
explode = [0.2, 0.04, 0.0, 0.2,0.0,0.2]
colors = ['saddlebrown', 'wheat', 'crimson', 'lightgrey','lightblue','darkblue']
plt.pie(data, 
        explode=explode, 
        labels=cars, 
        colors=colors, 
        autopct='%.2f%%',    # here we also add the % sign 
        shadow=True)
plt.show()
```

## The doughnut chart


 A doughnut chart is very similar to a pie chart. However, since a doughnut chart has an opening at the center, it makes slices look more like bars. 
 
 Unfortunately, there is no specific method for plotting a doughnut chart in the matplotlib library. But we can use the wedgeprops parameter to define the width of the wedges.

```python
plt.pie(data, 
        labels=cars, 
        colors=colors, 
        startangle=90, 
        wedgeprops={'width': 0.2})
plt.show()
```

The wedgeprops parameter also lets we change various pie slice features, such as linewidth, edgecolor, alpha (for transparency), and many others.

#### A histogram is like a bar chart that shows how often different numbers or values occur in a dataset. Instead of showing each number individually, it groups them into bins or ranges.

* Outlier Detection: To identify anomalies or outliers in datasets.
* Comparing Distributions: For comparing distributions between different datasets or groups.

# Box Plot

* Boxplot is popular graphical representation of given dataset statics
* Bottom line shows minimum values available in the dataset

#### The box has three things in it

* start of box show first qurantile 25%
* between the box mark shows 50%
* third/end of box shows 75%


Upper line show maximum value
three circles demonstrate outliers in the dataset: Outliers are the data points that are far away from centeric samples

```python
import numpy as np
import matplotlib.pyplot as plt

# Fixing random state for reproducibility
np.random.seed(19680801)

# fake up some data
spread = np.random.rand(50) * 100
center = np.ones(25) * 50
flier_high = np.random.rand(10) * 100 + 100
flier_low = np.random.rand(10) * -100
data = np.concatenate((spread, center, flier_high, flier_low))
```

```python
fig1, ax1 = plt.subplots()
ax1.set_title('Basic Plot')
ax1.boxplot(data)
```

```python
green_diamond = dict(markerfacecolor='g', marker='D')
fig3, ax3 = plt.subplots()
ax3.set_title('Changed Outlier Symbols')
ax3.boxplot(data, flierprops=green_diamond)
```

```python
red_square = dict(markerfacecolor='r', marker='s')
fig5, ax5 = plt.subplots()
ax5.set_title('Horizontal Boxes')
ax5.boxplot(data, vert=False, flierprops=red_square)
```

### Stack Plots

Stackplots are generated by plotting different datasets vertically on top of one another rather than overlapping with one another. Below we show some examples to accomplish this with Matplotlib.


```python
# importing modules
import numpy as np
import matplotlib.pyplot as plt

# values of x and y axes
x = [1, 2, 3, 4, 5]
y1 = [1, 1, 2, 3, 5]
y2 = [0, 4, 2, 6, 8]
y3 = [1, 3, 5, 7, 9]

# stacking the data vertically
y = np.vstack([y1, y2, y3])

# Adding lables
labels = ["Fibonacci ", "Evens", "Odds"]

fig, ax = plt.subplots()
ax.stackplot(x, y1, y2, y3, labels=labels)
ax.legend(loc='upper left')
plt.show()

fig, ax = plt.subplots()
ax.stackplot(x, y)

# function to show the plot
plt.show()
```

###  Stem Plots

matplotlib.pyplot.stem() creates stem plots.

A Stem plot plots vertical lines at each x position covered under the graph from the baseline to y, and places a marker there.

 Syntax: stem([x, ] y, linefmt=None, markerfmt=None, basefmt=None)

Parameters:



* x (array-like, optional): The x-positions of the stems. Default: (0, 1, …, len(y) – 1).
* y (array-like): The y-values of the stem heads.
* linefmt (str, optional): A string defining the properties of the vertical lines. Usually, this will be a color or a color and a linestyle:
* ‘-‘: solid line
* ‘–‘: dashed line
* ‘-.’: dash-dot line
* ‘:’ dotted line


```python
# importing libraries 
import matplotlib.pyplot as plt 
import numpy as np 

# values of x and y axes
x = np.linspace(0.1, 2 * np.pi, 41) 
y = np.exp(np.sin(x)) 

#Function for stem plot
plt.stem(x, y, use_line_collection = True) 

# function to show the plot
plt.show() 
```

### Default style
The default style is what we've been seeing so far throughout the book; we'll start with that. First, let's reset our runtime configuration to the notebook default: 

```python
# reset rcParams
IPython_default = plt.rcParams.copy()
plt.rcParams.update(IPython_default);
```

```python
def hist_and_lines():
    np.random.seed(0)
    fig, ax = plt.subplots(1, 2, figsize=(11, 4))
    ax[0].hist(np.random.randn(1000))
    for i in range(3):
        ax[1].plot(np.random.rand(10))
    ax[1].legend(['a', 'b', 'c'], loc='lower left')
```

```python
hist_and_lines()
```

### FiveThiryEight style

```python
with plt.style.context('fivethirtyeight'):
    hist_and_lines()
```
