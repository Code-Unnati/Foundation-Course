# Lab-2  Analyzing Flights Dataset with Seaborn

# Dataset:
Seaborn's built-in "flights" dataset.

# Objective: 

Explore the monthly passenger trends, visualize the correlation between years and months, and observe the changes in passenger numbers over the years.

# Steps:

## Import necessary libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights = sns.load_dataset("flights")
```

## Load the dataset

## Scatter Plot:

Create a scatter plot to visualize the monthly passenger trends over the years.

```python
# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x="year", y="passengers", data=flights, hue="month", palette="viridis")
plt.title("Monthly Passenger Trends Over Years")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.legend(title="Month")
plt.show()
```

### Smart Insights: Scatter Plot (Monthly Passenger Trends Over Years):

- The scatter plot illustrates the monthly passenger trends over the years.
- There is a general upward trend in the number of passengers over the years.
- Some months show higher passenger counts, indicating potential seasonal variations.

## Line Plot:
Plot the trend of passenger numbers over the years.

```python
# Create a line plot
plt.figure(figsize=(10, 6))
sns.lineplot(x="year", y="passengers", data=flights, ci=None)
plt.title("Passenger Trends Over Years")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.show()
```

### Smart Insights: Line Plot (Passenger Trends Over Years)

- The line plot depicts the trend of passenger numbers over the years.
- There is a consistent overall increase in the number of passengers from year to year.
- There is a noticeable upward trend with slight fluctuations.

# Violin Plot Grouped by Month:


```python


# Create a violin plot of the number of passengers grouped by month
sns.violinplot(x='month', y='passengers', data=flights, palette='coolwarm')
plt.title('Violin Plot of Number of Passengers by Month')
plt.xlabel('Month')
plt.ylabel('Number of Passengers')
plt.xticks(rotation=45)  # Rotate month labels for better readability
plt.show()
```

### Smart Insights: Violin Plot  

 The violin plot provides a comprehensive visualization of the distribution of the number of passengers for each month. Here are the key insights:

Seasonal Trends:

The plot reveals distinct seasonal patterns in air travel. For example, certain months like July and August show a higher density of passengers, indicating a peak travel season, likely due to summer vacations.

Distribution Shape:

Months like July and August have broader distributions, showing a wide range of passenger numbers, which might indicate varied travel schedules.


# Strip Plot Grouped by Year

```python
# Create a strip plot of the number of passengers grouped by year
sns.stripplot(x='year', y='passengers', data=flights, palette='coolwarm', jitter=True)
plt.title('Strip Plot of Number of Passengers by Year')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.show()
```

### Smart Insights: Violin Plot   

* Individual Observations:

The strip plot shows each individual observation, revealing the exact number of passengers for each flight within the dataset. This helps in identifying the specific data points and their distribution across the years.


* Comparative Analysis:

By comparing the distribution of points across different years, we can see changes in travel patterns and capacity. For instance, earlier years (e.g., 1949-1950) show fewer flights with high passenger numbers, whereas later years show more frequent high-capacity flights.
