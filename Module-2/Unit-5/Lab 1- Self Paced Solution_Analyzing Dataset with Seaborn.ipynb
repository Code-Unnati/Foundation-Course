## Count Plot:
Create a count plot to visualize the distribution of flights across months.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights = sns.load_dataset("flights")

# Create a count plot
plt.figure(figsize=(10, 6))
sns.countplot(x="month", data=flights, palette="viridis")
plt.title("Distribution of Flights by Month")
plt.xlabel("Month")
plt.ylabel("Number of Flights")
plt.show()
```

### Smart Insights: Countplot

- The count plot shows that the highest number of flights occurred in the month of August, while the lowest number was observed in February.

## Box Plot:
Generate a box plot to visualize the distribution of passenger counts across months and year.

```python
# Create a boxplot of the number of passengers grouped by month
sns.boxplot(x='month', y='passengers', data=flights, palette='coolwarm')
plt.title('Boxplot of Number of Passengers by Month')
plt.xlabel('Month')
plt.ylabel('Number of Passengers')
plt.xticks(rotation=45)  # Rotate month labels for better readability
plt.show()
```

```python
# Create a boxplot of the number of passengers grouped by year
sns.boxplot(x='year', y='passengers', data=flights, palette='coolwarm')
plt.title('Boxplot of Number of Passengers by Year')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.show()
```

### Smart Insights: Box Plot

- These boxplots will provide a clear visualization of how the number of passengers varies over time, both annually and monthly.

## Hist Plot:
Create a Hist plot to better understand the distribution of passenger counts across frequency.

```python
# Create a histogram of the number of passengers
sns.histplot(
    flights['passengers'], 
    bins=20, 
    kde=True, 
    color='skyblue', 
    edgecolor='black', 
)
plt.title('Distribution of Passengers')
plt.xlabel('Number of Passengers')
plt.ylabel('Frequency')
plt.show()
```

### Smart Insights: Hist Plot

* It provides a clearer and more detailed representation of the distribution of passengers in the 'flights' dataset.
