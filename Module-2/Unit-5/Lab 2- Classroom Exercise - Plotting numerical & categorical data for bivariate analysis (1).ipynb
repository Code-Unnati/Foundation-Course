# Lab - 2  Plotting numerical & categorical data for bivariate analysis

### ➡️ Import Libraries


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

#### ➡️ Read dataset


```python
df = pd.read_csv("/content/Automobile_data.csv")
```

#  | Scatter Plot |

##### Importance of Scatter Plots in Seaborn


* Visualizing Relationships: Displays the relationship between two continuous variables.


* Detecting Outliers: Outliers can be easily identified as points that deviate from the trend.

* Identifying Trends:  Reveals linear or non-linear relationships and patterns.

* Highlighting Clusters: Helps identify clusters or groupings within the data.


#### ➡️ official documentation   :-  https://seaborn.pydata.org/generated/seaborn.scatterplot.html

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x='engine-size',y='wheel-base',data=df,color='#FF0000',marker='>')
plt.show()
```

```python
#Changing the color,marker and figure size  of data points
plt.figure(figsize=(10,7))
sns.scatterplot(x='engine-size',y='wheel-base',data=df,marker='>',color='#FF0000',hue='fuel-type',style='num-of-cylinders')
plt.show()
```

```python
# Show groups with different colors using "hue" and tip size with "size  "
plt.figure(figsize=(10,7))
sns.scatterplot(x='engine-size',y='wheel-base',hue='fuel-type',size='city-mpg',sizes=(20,200),data=df)
plt.show()
```

# | Line plot |

#### Importance of Line Plots in Seaborn

* Trend Visualization: Visualizes trends over time or continuous intervals clearly.

* Built-in Aesthetics and Customization:Offers attractive themes, color palettes, and easy customization.

* Statistical Estimation: Computes and displays confidence intervals automatically.

* Integration with Pandas and Faceting: Integrates with Pandas and supports faceting for comparing data subsets.

#### ➡️ official documentation   :-  https://seaborn.pydata.org/generated/seaborn.lineplot.html

```python
# Plotting lineplot using sns.lineplot()
plt.style.use('seaborn-darkgrid')
%matplotlib inline


sns.lineplot(x='engine-size',y='wheel-base',data=df)
plt.show()
```

```python
""" - Adjusting background color using axes.facecolor
- Changing label size using xtick.labelsize , ytick.labelsize """


plt.figure(figsize=(14,6))
sns.set(rc={'axes.facecolor':'cornflowerblue',"axes.grid":True,'xtick.labelsize':14,'ytick.labelsize':14})
sns.lineplot(x='engine-size',y='wheel-base',data=df , linewidth = 2.5)
sns.lineplot(x='engine-size',y='wheel-base',data=df , linewidth = 2.5)
plt.show()
```

## | Regplot |

### Regplot

* Visualizing Linear Relationships: Specifically shows linear relationships between two variables.

* Adding Regression Line: Automatically adds a regression line to the scatter plot.

* Assessing Fit Quality: Helps assess the quality and strength of the linear fit.

* Displaying Confidence Intervals: Shows confidence intervals around the regression line.

#### ➡️ official documentation   :-  https://seaborn.pydata.org/generated/seaborn.regplot.html

```python
#Fit the regression model using log(x)
plt.figure(figsize=(9,9))
sns.regplot(x=df["compression-ratio"] , y=df["engine-size"] , color='#FF6600' , logx=True)
plt.show()
```

```python
# Use 50% Confidnce interval
plt.figure(figsize=(9,9))
sns.regplot(x=df["compression-ratio"] , y=df["engine-size"] , color='#FF6600' , ci=50 , truncate=False)
plt.show()
```

## | Violinplot |
A violin plot plays a similar role as a box and whisker plot. It shows the distribution of quantitative data across several levels of one (or more) categorical variables such that those distributions can be compared.

Unlike a box plot, in which all of the plot components correspond to actual datapoints, the violin plot features a kernel density estimation of the underlying distribution.

#### ➡️ official documentation   :-  https://seaborn.pydata.org/generated/seaborn.violinplot.html

```python
%matplotlib inline
sns.set_style("white")
```

```python
#Scale the density relative to the counts across all bins
plt.figure(figsize=(16,7))
sns.violinplot(x=df["body-style"] , y = df["city-mpg"] , palette="Set1" , scale="count")
plt.show()
```

```python
#Show groups with different colors using "hue" and "palette"
plt.figure(figsize=(16,7))
sns.violinplot(x=df["body-style"] , y = df["city-mpg"] , palette="Set1", hue= df["fuel-type"])
plt.show()
```

```python
#Show groups with different colors using "hue" and "palette"
# Displaying multiple violin plots using subplot function.
fig1 , axes = plt.subplots(nrows=2,ncols=2 , figsize = (20,20))
sns.violinplot(x="body-style" , y = "city-mpg" , ax = axes[0,0] ,data=df , inner=None)
sns.violinplot(x="body-style" , y = "wheel-base" ,ax = axes[0,1] , data=df , inner=None)
sns.violinplot(x="body-style" , y = "length" , ax = axes[1,0] , data=df, inner=None)
sns.violinplot(x="body-style" , y = "width" , ax = axes[1,1] , data=df, inner=None )
plt.show()
```

# | Stripplot and Swarmplot |


The stripplot will draw a scatterplot where one variable is categorical. A strip plot can be drawn on its own, but it is also a good complement to a box or violin plot in cases where you want to show all observations along with some representation of the underlying distribution.

The swarmplot is similar to stripplot(), but the points are adjusted (only along the categorical axis) so that they don’t overlap. This gives a better representation of the distribution of values, although it does not scale as well to large numbers of observations (both in terms of the ability to show all the points and in terms of the computation needed to arrange them).

#### ➡️ official documentation   :-  https://seaborn.pydata.org/generated/seaborn.stripplot.html

#### ➡️ official documentation   :-  https://seaborn.pydata.org/generated/seaborn.swarmplot.html

```python
# Defining the size of the plots
plt.figure(figsize=(7,7))
sns.swarmplot([df['fuel-type'], df['city-mpg']], size=10)
plt.show()
```

```python
# Displaying multiple swarmplots using subplot function.
fig1 , axes = plt.subplots(nrows=2,ncols=2 , figsize = (20,20))
sns.swarmplot([df['fuel-type'],df['city-mpg']], ax = axes[0,0], size=8)
sns.swarmplot([df['fuel-type'] , df['wheel-base']] ,ax = axes[0,1], size=8)
sns.swarmplot([df['fuel-type'] ,df['length']] , ax = axes[1,0] , size=8)
sns.swarmplot([df['fuel-type'] , df['width']] , ax = axes[1,1] , size=8)
plt.show()
```

```python
#Changing edge color , size and linewidth of data points
plt.figure(figsize=(7,7))
sns.swarmplot(x= "fuel-type", y = "wheel-base", size = 9 , linewidth= 2 , edgecolor="black" , data =df)
plt.show()
```

```python
""" "Jitter" parameter signifies the amount of jitter to apply.
This can be extremely useful when we have large clusters of data points"""

plt.figure(figsize=(11,8))
sns.stripplot(x=df["body-style"] ,palette="Set1", y = df["engine-size"],jitter=.3)
plt.show()
```

```python
# Adjust the linewidth of the edges of the circles using "linewidth" parameter
# Adjust the size of the circles using the "size" parameter
plt.figure(figsize=(14,10))
sns.stripplot(x=df["body-style"] ,palette="Set1", y = df["engine-size"],linewidth=2,size=8)
plt.show()
```

```python
plt.figure(figsize=(16,10))
sns.stripplot(x=df["body-style"] ,palette="Set2", y = df["engine-size"],hue=df["fuel-type"],marker = "o",size=30,alpha = .15)
plt.show()
```

```python
# Drawing stripplot on top of a box plot

plt.figure(figsize=(11,8))
sns.stripplot(x=df["body-style"] ,palette="Set1", y = df["engine-size"],hue=df["fuel-type"],jitter=True, color="black",linewidth=2,size=8)
sns.boxplot(x=df["body-style"] ,palette="Set1", y = df["engine-size"] , color='black')
plt.show()
```

```python
# Drawing stripplot on top of a violin plot
plt.figure(figsize=(14,10))
sns.stripplot(x=df["body-style"] ,palette="Set1", y = df["engine-size"],hue=df["fuel-type"],jitter=True, color="black",linewidth=2,size=8)
sns.violinplot(x=df["body-style"] ,palette="Set1", y = df["engine-size"],hue=df["fuel-type"],scale="count")
plt.show()
```
