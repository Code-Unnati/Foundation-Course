# Lab - 3 Analyzing Flights Dataset with Seaborn

# Dataset:
Seaborn's built-in "flights" dataset.

# Objective: 

Visualize the distribution of continuous numerical data.

# Steps:

## Import necessary libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

## Load the dataset

```python
# Load the flights dataset
flights = sns.load_dataset("flights")
```

## Instructions:
Writedown the pointswise insights below all the charts respectively.

```python
# Create a pairplot with 'hue' parameter
sns.pairplot(flights, hue='month')
plt.show()
```

## smart insight:

* Diagonal Plots (Histograms):

Along the diagonal of the pairplot, histograms are displayed for each numerical variable ('year' and 'passengers').

The histograms show the distribution of values for each variable separately.

* Off-Diagonal Plots (Scatterplots):

The off-diagonal plots show scatterplots of 'year' vs. 'passengers', exploring the relationship between these variables.

Since both 'year' and 'passengers' are numerical variables, each point represents a specific combination of year and passenger count

# Heatmap:
Generate a heatmap to show the correlation between years and months.

```python
# Pivot the data to create a matrix suitable for a heatmap
flights_pivot = flights.pivot_table(index="month", columns="year", values="passengers")

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(flights_pivot, cmap="YlGnBu", annot=True, fmt="d")
plt.title("Passenger Count by Year and Month")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```

## smart insight:

* Seasonal Trends:

The heatmap illustrates seasonal patterns in air travel over the years. Darker shades represent higher passenger counts, while lighter shades indicate lower counts.

* Yearly Variations:

By comparing the intensity of colors across different years for each month, we can identify variations in passenger counts over time.

###  Jointplot:
Generate a jointplot to visualize the relationship between 'year' and 'passengers'.

```python
# Create a jointplot with hue parameter
sns.jointplot(x='year', y='passengers', data=flights, kind='scatter', cmap='YlGnBu',  hue='month')
plt.title('Jointplot of Year vs. Passengers with Month Hue')
plt.xlabel('Year')
plt.ylabel('Passengers')
plt.show()
```

## smart insight:

* Comparative Analysis:

By comparing the color distribution across different months, we can identify variations in air travel demand throughout the year. This comparison helps in understanding the relative popularity of different months for air travel.

* Seasonal Patterns:

Darker shades of colors may indicate months with higher passenger counts, suggesting peak travel seasons, while lighter shades may indicate off-peak periods.
