# Lab 1 Linked Lists in Python

Linked lists have several use cases in computer science and programming due to their flexibility and efficient memory management. Here are some common use cases for linked lists:
<ul>
    <li><b>Dynamic Memory Allocation:</b> Linked lists allow for dynamic memory allocation, meaning that memory can be allocated and deallocated at runtime. This is particularly useful when the size of the data structure is not known in advance or when memory needs to be managed efficiently.</li><br>
    <li><b>Implementation of Stacks and Queues:</b> Linked lists are the underlying data structure for implementing stacks and queues due to their ability to efficiently add and remove elements from either end of the list (for stacks: LIFO - Last In, First Out; for queues: FIFO - First In, First Out).</li><br>
    <li><b>Dynamic Data Structures:</b> Unlike arrays, linked lists can easily grow or shrink in size during program execution. This makes linked lists suitable for applications where the size of the data structure needs to change dynamically.</li><br>
    <li><b>Efficient Insertions and Deletions:</b> Insertions and deletions at arbitrary positions within the list can be done efficiently in linked lists with constant time complexity (O(1)), as long as you have a reference to the node where you want to perform the operation.</li><br>
    <li><b>Memory Efficiency:</b> Linked lists are memory-efficient compared to arrays when dealing with frequent insertions and deletions, as they do not require contiguous blocks of memory. Each node in a linked list only requires memory for the data it contains and a reference to the next node.</li><br>
    <li><b>Implementation of Graphs:</b> Linked lists can be used to represent adjacency lists in graph data structures, where each node in the linked list represents a vertex in the graph, and the linked list contains the adjacent vertices.</li><br>
    <li><b>Arbitrary Memory Allocation:</b> Linked lists can be used to manage memory for objects of varying sizes, such as in memory allocators or garbage collectors.</li><br>
</ul>
Overall, linked lists are versatile data structures that find applications in various domains, including operating systems, database systems, text editors, and more
