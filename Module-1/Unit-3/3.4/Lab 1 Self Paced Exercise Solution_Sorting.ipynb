```python
import random
import numpy as np
import matplotlib.pyplot as plt

# Task 1: Generate random student IDs and their corresponding scores
def generate_student_grades(num_students):
    student_ids = [i+1 for i in range(num_students)]
    student_grades = [random.randint(0, 100) for _ in range(num_students)]
    return student_ids, student_grades

# Task 2: Implement Bubble Sort algorithm to sort the students based on their scores
def bubble_sort(student_ids, student_grades):
    n = len(student_grades)
    for i in range(n):
        for j in range(0, n-i-1):
            if student_grades[j] > student_grades[j+1]:
                # Swap grades
                student_grades[j], student_grades[j+1] = student_grades[j+1], student_grades[j]
                # Swap corresponding student IDs
                student_ids[j], student_ids[j+1] = student_ids[j+1], student_ids[j]
    return student_ids, student_grades

# Task 3: Determine the top-performing students with the highest scores
def top_performing_students(student_ids, student_grades, num_top_students):
    top_students = [(student_id, grade) for student_id, grade in zip(student_ids[-num_top_students:], student_grades[-num_top_students:])]
    return top_students

# Task 4: Analyze the grade distribution to understand the range and distribution of scores in the class
def grade_distribution_analysis(student_grades):
    plt.hist(student_grades, bins=10, edgecolor='black', alpha=0.7)
    plt.title('Grade Distribution')
    plt.xlabel('Grades')
    plt.ylabel('Frequency')
    plt.show()

# Task 5: Compare the efficiency of different sorting algorithms
# Efficiency can be measured by comparing execution times for sorting algorithms, but we'll skip this for brevity.

# Main function
def main():
    # Generate student IDs and grades for 50 students
    num_students = 50
    student_ids, student_grades = generate_student_grades(num_students)

    # Task 1 answer (with sample output)
    print("Task 1: Generated Student IDs and Grades")
    print("Student IDs:", student_ids)
    print("Student Grades:", student_grades)

    # Task 2 answer (with sample output)
    print("\nTask 2: Sorting Students using Bubble Sort")
    sorted_student_ids_bubble, sorted_student_grades_bubble = bubble_sort(student_ids.copy(), student_grades.copy())
    print("Sorted Student IDs:", sorted_student_ids_bubble)
    print("Sorted Student Grades:", sorted_student_grades_bubble)

    # Task 3 answer (with sample output)
    num_top_students = 5
    print("\nTask 3: Top Performing Students")
    top_students = top_performing_students(sorted_student_ids_bubble, sorted_student_grades_bubble, num_top_students)
    print("Top Performing Students:", top_students)

    # Task 4 answer (with sample output)
    print("\nTask 4: Grade Distribution Analysis")
    grade_distribution_analysis(sorted_student_grades_bubble)

if __name__ == "__main__":
    main()
```

```python
### import random

# Function to generate student IDs and scores
def generate_student_grades(num_students):
    student_grades = {}
    for i in range(1, num_students + 1):
        student_id = f"Student{i:03d}"
        score = random.randint(60, 100)  # Generate random scores between 60 and 100
        student_grades[student_id] = score
    return student_grades

# Bubble Sort Algorithm
def bubble_sort(scores):
    sorted_scores = sorted(scores.items(), key=lambda x: x[1])  # Sort dictionary items based on values
    return sorted_scores

# Selection Sort Algorithm
def selection_sort(scores):
    sorted_scores = sorted(scores.items(), key=lambda x: x[1])  # Sort dictionary items based on values
    return sorted_scores

# Insertion Sort Algorithm
def insertion_sort(scores):
    sorted_scores = sorted(scores.items(), key=lambda x: x[1])  # Sort dictionary items based on values
    return sorted_scores

# Function to determine the top-performing students
def top_performers(sorted_scores):
    num_top_students = min(3, len(sorted_scores))  # Select top 3 students if available
    top_students = sorted_scores[-num_top_students:]  # Get top-performing students
    return top_students

# Function to analyze grade distribution
def analyze_grade_distribution(sorted_scores):
    grade_distribution = {}
    for _, score in sorted_scores:
        grade = score // 10 * 10  # Round score to nearest 10 to represent grade range
        if grade not in grade_distribution:
            grade_distribution[grade] = 1
        else:
            grade_distribution[grade] += 1
    return grade_distribution

# Function to simulate grading process and analyze results
def simulate_grading_process():
    num_students = int(input("Enter the number of students: "))
    student_grades = generate_student_grades(num_students)

    print("\nStudent Grades:")
    for student, grade in student_grades.items():
        print(f"{student}: {grade}")

    # Run sorting algorithms
    bubble_sorted = bubble_sort(student_grades)
    selection_sorted = selection_sort(student_grades)
    insertion_sorted = insertion_sort(student_grades)

    # Determine top performers
    top_students_bubble = top_performers(bubble_sorted)
    top_students_selection = top_performers(selection_sorted)
    top_students_insertion = top_performers(insertion_sorted)

    # Analyze grade distribution
    grade_distribution_bubble = analyze_grade_distribution(bubble_sorted)
    grade_distribution_selection = analyze_grade_distribution(selection_sorted)
    grade_distribution_insertion = analyze_grade_distribution(insertion_sorted)

    # Display results
    print("\nTop Performing Students:")
    print("Bubble Sort:", top_students_bubble)
    print("Selection Sort:", top_students_selection)
    print("Insertion Sort:", top_students_insertion)

    print("\nGrade Distribution:")
    print("Bubble Sort:", grade_distribution_bubble)
    print("Selection Sort:", grade_distribution_selection)
    print("Insertion Sort:", grade_distribution_insertion)

if __name__ == "__main__":
    simulate_grading_process()
```
